```python
import cellular_automata as ca
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
```

```python
GENERATIONS = 100
ROWS = COLS = 101
ORIGIN = (COLS // 2, ROWS // 2)
PHI = 1000
CANCER_INIT_POSITIONS = [(ORIGIN[0], ORIGIN[1]), (ORIGIN[0] + 1, ORIGIN[1]), 
                         (ORIGIN[0] - 1, ORIGIN[1]), (ORIGIN[0], ORIGIN[1] - 1), 
                         (ORIGIN[0], ORIGIN[1] + 1)]
# K1, K2, K3, K4 = 0.74, 0.2, 0.4, 0.4
RHO = 3.85
TAU = 7
```

## Varying k1

```python
length = 5
k1s = np.linspace(0.20,0.89,length)
k2s = np.full(length,0.2)
k34s = np.full(length,0.4)
```

```python
histories = []
for k1,k2,k3,k4 in zip(k1s,k2s,k34s,k34s):
    histories.append(ca.simulate_tumor_growth(TAU,GENERATIONS, ROWS, COLS, PHI, RHO, k1, k2, k3, k4, CANCER_INIT_POSITIONS, ORIGIN,))
```

```python
k1s = np.linspace(0.20,0.89,length)


fig, ax = plt.subplots(1,3, figsize=[15,3])
for i,k in enumerate(k1s):
    ax[0].plot([histories[i][g]['Nc'] for g in range(GENERATIONS)], label=k)
    ax[1].plot([histories[i][g]['Ne'] for g in range(GENERATIONS)], label=k)
    ax[2].plot([histories[i][g]['Nd'] for g in range(GENERATIONS)], label=k)
ax[0].set_title("Cancerous")
ax[1].set_title("Effector")
ax[2].set_title("Diseased")
ax[2].legend()
fig.suptitle("Varying k1")
# .ylabel("Number of cells")
```
