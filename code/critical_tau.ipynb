```python
import cellular_automata as ca
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
import numpy as np
```

```python
GENERATIONS = 200
ROWS = COLS = 101
ORIGIN = (COLS // 2, ROWS // 2)
PHI = 1000
CANCER_INIT_POSITIONS = [(ORIGIN[0], ORIGIN[1]), (ORIGIN[0] + 1, ORIGIN[1]), 
                         (ORIGIN[0] - 1, ORIGIN[1]), (ORIGIN[0], ORIGIN[1] - 1), 
                         (ORIGIN[0], ORIGIN[1] + 1)]
K1, K2, K3, K4 = 0.74, 0.2, 0.4, 0.4
RHO = 3.85
# TAU = 7
```

```python
tau_histories = []

for tau in range(25,45):
    tau_histories.append(ca.simulate_tumor_growth(tau, GENERATIONS, ROWS, COLS, PHI, RHO, K1, K2, K3, K4, CANCER_INIT_POSITIONS, ORIGIN))
```

```python
# k1s = np.linspace(0.20,0.89,length)
taus = [i for i in range(25,45)]
palette = sns.color_palette('rocket_r', len(taus))

fig, ax = plt.subplots(1,3, figsize=[20,4],)
for i,tau in enumerate(taus):
    ax[0].plot([tau_histories[i][g]['Nc'] for g in range(GENERATIONS)], color=palette[i], label=tau)
    ax[1].plot([tau_histories[i][g]['Ne'] for g in range(GENERATIONS)], color=palette[i], label=tau)
    ax[2].plot([tau_histories[i][g]['Nd'] for g in range(GENERATIONS)], color=palette[i], label=tau)
ax[0].set_title("Cancerous")
ax[1].set_title("Effector")
ax[2].set_title("Diseased")
ax[2].legend(loc='upper right', bbox_to_anchor=(1.5, 1))
# fig.suptitle(r"Varying $\tau$")
# .ylabel("Number of cells")
```
